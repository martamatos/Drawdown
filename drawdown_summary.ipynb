```python
import pandas as pd
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
data = pd.read_csv('Summary of Solutions by Overall Rank  Drawdown.csv', index_col=0)
```

```python
data['Net Cost (BILLIONS US $)'] = data['Net Cost (BILLIONS US $)'].str.replace('$', '')
data['Savings (BILLIONS US $)'] = data['Savings (BILLIONS US $)'].str.replace('$', '')
data['Net Cost (BILLIONS US $)'] = data['Net Cost (BILLIONS US $)'].str.replace(',', '')
data['Savings (BILLIONS US $)'] = data['Savings (BILLIONS US $)'].str.replace(',', '')
data['Net Cost (BILLIONS US $)'] = data['Net Cost (BILLIONS US $)'].str.replace('.', '')
data['Savings (BILLIONS US $)'] = data['Savings (BILLIONS US $)'].str.replace('.', '')
data['Net Cost (BILLIONS US $)']= pd.to_numeric(data['Net Cost (BILLIONS US $)'])
data['Savings (BILLIONS US $)']= pd.to_numeric(data['Savings (BILLIONS US $)'])
data.head()
```

```python
data['savings over cost'] = data['Savings (BILLIONS US $)'] / data['Net Cost (BILLIONS US $)']
```

```python
data['savings over cost']=data['savings over cost'].round(1)
```

```python
data = data[:-1]
```

```python
fig, ax = plt.subplots()
ax.hist(data['TOTAL ATMOSPHERIC CO2-EQ REDUCTION (GT)'].values, bins=100);
ax.set_ylabel('TOTAL ATMOSPHERIC CO2-EQ REDUCTION (GT)')
ax.set_xlim([0,30]);
```

```python
data['ATMOSPHERIC CO2-EQ REDUCTION (% of total)'] = data['TOTAL ATMOSPHERIC CO2-EQ REDUCTION (GT)'] / data['TOTAL ATMOSPHERIC CO2-EQ REDUCTION (GT)'].sum() * 100
```

```python
data = data[['Solution', 'Sector', 'TOTAL ATMOSPHERIC CO2-EQ REDUCTION (GT)',
            'ATMOSPHERIC CO2-EQ REDUCTION (% of total)',
            'Net Cost (BILLIONS US $)', 'Savings (BILLIONS US $)',
            'savings over cost']]
```
